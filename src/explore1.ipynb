```python
linreq_model=LinearRegression()
linreq_model.fit(x_train,y_train)
dt_model=DecisionTreeClassifier(max_depth=5)
dt_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
```

```python
from sklearn.metrics import accuracy_score,classification_report,precision_score,recall_score,confusion_matrix
```

accuracy_score= modelin toplamda ne kadar doğru yaptığını

precision= pozitif dediği kişilerin yüzde kaçı gerçekten pozitif (Model POZİTİF dediğinde ne kadar doğru?) false positive önemliyse “Pozitif dediysem gerçekten pozitif olmalı.”


recall =Model GERÇEK POZİTİFLERİ ne kadar yakaladı? False negative önemliyse hatalarının çok pahalı olduğu durumlarda.
“Pozitifleri kaçırmamalıyım.”

Harmonik Ortalama → “küçüğü cezalandırır”
F1 Skoru → “Precision & Recall ikisi de iyi mi?”

Precision yüksek olabilir ama recall çok düşükse → kötü model
Recall yüksek olabilir ama precision çok düşükse → kötü model
F1 bunların ikisini dengeler ve gerçek kaliteyi gösterir.

confussion matrix=                Tahmin: 0     Tahmin: 1
                    Gerçek 0   →      TN            FP
                    Gerçek 1   →      FN            TP


```python
def evulation(x_matrix,y_target,model):
    preds=model.predict(x_matrix)
    print(f"\n\n on the data")
    print(f"mean_absolute_error :{mean_absolute_error(y_true=y_target,y_pred=preds)}")
    print(f"mean_squared_error :{mean_squared_error(y_true=y_target,y_pred=preds)}")
    print(f"_root_mean_squared_error : {mean_absolute_error(y_true=y_target,y_pred=preds)}")
    print(f"r2_score : {r2_score(y_true=y_target,y_pred=preds)}")
    
```

```python
def tree_evulation(x_matrix,y_target,model):
    preds=model.predict(x_matrix)
    print(f"\n\n on the data")
    print(f"accuracy_score : {accuracy_score(y_true=y_target,y_pred=preds)}")
    print(f"recall_score : {recall_score(y_true=y_target,y_pred=preds)}")
    print(f"precision_score : {precision_score(y_true=y_target,y_pred=preds)}")
    print(f"confussion_matrix :\n {confusion_matrix(y_true=y_target,y_pred=preds)}")
```

```python

evulation(x_train,y_train,linreq_model)
evulation(x_test,y_test,linreq_model)


```

```python
tree_evulation(x_train,y_train,dt_model)
tree_evulation(x_test,y_test,dt_model)
```

```python
hyperparameters_dict={
    'criterion' : ['gini','entropy','log_loss'],
    'max_depth' : [2,5,10,20],
    'min_samples_split' : [2,5,10,15,20],
    'min_samples_leaf' : [2,5,10]
}
```

Derinlik + min_samples → ağacın büyüklüğünü kontrol eder

Criterion → hangi kural ile bölünecek

max_features → hesaplama hızını etkiler

class_weight → dengesiz veri varsa önemli Hiperparametre	Ne yapıyor?	

max_depth	Ağacın derinliği	Çok derin → ezberler (overfitting), çok sığ → öğrenemez (underfitting)

min_samples_split	Düğümü bölebilmek için gereken minimum örnek	Küçük → çok bölünür, büyük → az bölünür

min_samples_leaf	Yaprak düğümde minimum örnek	Küçük → overfitting, büyük → basit model

criterion	Hangi ölçüte göre bölünecek	'gini' veya 'entropy' genelde yeterli

gini → “Düğümde karışıklık ne kadar?”

Hedef: tek bir sınıf olsun.

Hızlı ve yaygın.

entropy → “Düğümde bilgi kazancı ne kadar?”

Hedef: düğümü daha düzenli yap.

max_depth = Ağacın derinliği

Ağaç = Karar ağacı, her düğüm bir soru.

Derinlik = soruların üst üste kaç katman olabileceği.

min_samples_split = Bir düğümün bölünebilmesi için gereken minimum örnek sayısı

Ağacın bir düğümü daha fazla dallara ayırabilmesi için en az kaç örnek olmalı, onu belirler. örnekten kastı satır

min_samples_leaf → her yaprakta en az kaç örnek olmalı 

min_samples_leaf = 2

Bölünürken oluşacak her yaprakta en az 2 örnek olmalı


```python
from sklearn.model_selection import GridSearchCV
```

GridSearchCV = “Otomatik deneme-yanılma”

Modelin ayarlarını (hyperparameter’ları) en iyi şekilde bulmak için kullanılır.

Yani en iyi kombinasyonu otomatik dener ve seçer.

```python
gridsearch_model=GridSearchCV(estimator=DecisionTreeClassifier(),
                             param_grid=hyperparameters_dict,
                              scoring="accuracy",
                              verbose=3 )
```

```python
gridsearch_model.fit(x_train,y_train)
```

estimator=DecisionTreeClassifier()

Hangi modeli optimize edeceğini söylüyor.

Burada Decision Tree kullanıyoruz.

param_grid=hyperparameters_dict

Denenecek parametreleri veriyoruz.

Örneğin: max_depth, min_samples_split, criterion gibi.

GridSearchCV bunların tüm kombinasyonlarını dener.

scoring='f1'

Modeli hangi ölçüte göre değerlendireceğini söylüyor.

Burada F1 score kullanıyoruz (Precision ve Recall’u dengeler).

verbose=3

Çalışırken ekrana detaylı bilgi yazdırır.

Kaç kombinasyon denendiğini görebilirsin.

```python
tree_evulation(x_train,y_train,gridsearch_model)
tree_evulation(x_test,y_test,gridsearch_model)
```
